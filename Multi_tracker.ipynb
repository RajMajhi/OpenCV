```python
# Multiple object tracking
import cv2

trackers = cv2.legacy.MultiTracker_create()

v = cv2.VideoCapture(r"C:\Users\rajm4\Downloads\data.mp4")
ret, frame = v.read()

# ---------- make window small ----------
cv2.namedWindow("Frame", cv2.WINDOW_NORMAL)
cv2.resizeWindow("Frame", 800, 500)

# ---------- ROI selection ----------
boxes = []

while True:
    cv2.imshow("Frame", frame)

    box = cv2.selectROI("Frame", frame)

    # if user presses ENTER without selecting → stop
    if box[2] == 0 or box[3] == 0:
        print("Selection finished")
        break

    boxes.append(box)
    print("Object added:", box)

# ---------- add trackers ----------
for box in boxes:
    tracker = cv2.legacy.TrackerCSRT_create()
    trackers.add(tracker, frame, box)

print("Tracking started")
print("Press Q → quit")
print("Press S → screenshot")

# ---------- tracking loop ----------
while True:
    ret, frame = v.read()
    if not ret:
        break

    success, boxes = trackers.update(frame)

    for box in boxes:
        x, y, w, h = [int(i) for i in box]
        cv2.rectangle(frame, (x,y), (x+w,y+h), (0,255,0), 2)

    cv2.imshow("Frame", frame)

    key = cv2.waitKey(30) & 0xFF

    # quit
    if key == ord('q'):
        print("Exiting...")
        break

    # screenshot
    if key == ord('s'):
        cv2.imwrite("screenshot.png", frame)
        print("Screenshot saved")

v.release()
cv2.destroyAllWindows()
```
