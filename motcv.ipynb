```python
# OpenCV Single Object Trackers (for testing only)
# Note: These are NOT used in MOTIP/SORT pipelines.
# They are useful for learning tracking behavior and debugging video input.

# CSRT
# Most accurate classical tracker in OpenCV.
# Handles scale change, rotation, partial occlusion well.
# Uses correlation filters with channel & spatial reliability.
# Slowest among traditional trackers.
# Good when object motion is fast or irregular.
# Recommended for experiments where accuracy > speed.
# tracker = cv2.TrackerCSRT_create()

# KCF
# Faster than CSRT but slightly less accurate.
# Uses kernelized correlation filters.
# Struggles with heavy occlusion or large scale changes.
# Good trade-off between speed and accuracy.
# tracker = cv2.TrackerKCF_create()

# BOOSTING
# One of the oldest trackers.
# Based on AdaBoost classifier.
# Very sensitive to noise and drift.
# Rarely used in modern work.
# tracker = cv2.TrackerBoosting_create()

# MIL (Multiple Instance Learning)
# Improves over Boosting by handling ambiguity in samples.
# More stable but still drifts over time.
# Medium speed, medium accuracy.
# tracker = cv2.TrackerMIL_create()

# TLD (Tracking-Learning-Detection)
# Attempts long-term tracking.
# Can re-detect object after losing it.
# Very unstable in practice and slow.
# Often avoided in research setups.
# tracker = cv2.TrackerTLD_create()

# MEDIANFLOW
# Works well only for slow, smooth motion.
# Fails if object moves quickly or disappears.
# Good for small predictable motion.
# tracker = cv2.TrackerMedianFlow_create()

# MOSSE
# Extremely fast (hundreds of FPS).
# Very low accuracy.
# Breaks with scale/rotation changes.
# Useful for real-time demos, not research.
# tracker = cv2.TrackerMOSSE_create()

# ===== Recommended choice for testing =====
# tracker = cv2.TrackerCSRT_create()   . Best accuracy but slower
```

```python
#for single tracking 


    # 'kcf': cv2.TrackerKCF_create, 
    # 'boosting': cv2.TrackerBoosting_create, 
    # 'mil': cv2.TrackerMIL_create,
    # 'tld': cv2.TrackerTLD_create,
    # 'medianflow': cv2.TrackerMedianFlow_create,
    # 'mosse': cv2.TrackerMOSSE_create 


import cv2

tracker = cv2.TrackerCSRT_create()

v = cv2.VideoCapture(r"C:\Users\rajm4\Downloads\data.mp4")
ret, frame = v.read()

cv2.namedWindow("Frame", cv2.WINDOW_NORMAL)
cv2.resizeWindow("Frame", 800, 500)

cv2.imshow("Frame", frame)
bb = cv2.selectROI("Frame", frame)

tracker.init(frame, bb)

while True:
    ret, frame = v.read()
    if not ret:
        break

    ok, bb = tracker.update(frame)

    if ok:
        x,y,w,h = [int(i) for i in bb]
        cv2.rectangle(frame,(x,y),(x+w,y+h),(0,255,0),2)

    cv2.imshow("Frame", frame)

    key = cv2.waitKey(30) & 0xFF
    if key == ord('q'):
        break
    if key == ord('s'):
        cv2.imwrite("shot.png", frame)
        print("saved")

v.release()
cv2.destroyAllWindows()
```
